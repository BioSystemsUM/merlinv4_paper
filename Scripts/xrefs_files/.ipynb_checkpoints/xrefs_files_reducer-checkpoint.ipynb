# Metabolites mapper reduction

```python
import pandas as pd

df = pd.read_csv("/Users/anabarbosa/Downloads/chem_xref.tsv", sep="\t", skiprows=351)
```

```python
df
```

```python
df = df.drop(["description"], axis=1)
```

```python
df['#source'].unique()
```

```python
bigg_compounds = df[df['#source'].str.count('bigg\.metabolite\:.*')>0]
```

```python
bigg_compounds
```

```python
kegg_compounds = df[df['#source'].str.count('kegg\.compound\:.*')>0]
```

```python
kegg_compounds
```

```python
seed_compounds = df[df['#source'].str.count('seed\.compound\:.*')>0]
```

```python
seed_compounds
```

```python
metacyc_compounds = df[df['#source'].str.count('metacyc\.compound\:.*')>0]
```

```python
metacyc_compounds
```

```python
all_dbs_compounds = bigg_compounds.append(kegg_compounds).append(seed_compounds).append(metacyc_compounds)
```

```python
all_dbs_compounds
```

```python
all_dbs_compounds.to_csv("chem_xrefs_reduced.csv",index=False)
```
